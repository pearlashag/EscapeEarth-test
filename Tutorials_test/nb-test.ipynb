# create DT classifier

```python
#drop nans ----- b/c DTs need numerical data
table1 = table1.dropna()
table2 = table2.dropna()
len(table1),len(table2) #compare to table lens above
```

```python

# make an array with binaries values in the len of each df

valid = np.repeat(1,len(table1))
invalid = np.repeat(0,len(table2))  
classes = np.append(valid,invalid)

#np.zeros(len(table2))---didnt like b/c float
# print(len(valid),len(invalid),valid[0:5],invalid[0:5],len(classes),classes[34027:34033])
```

# Start of Next Session

```python
#join the dfs

# tables = table1.append(table2)#, axis=0)
tables = pd.concat([table1,table2], axis=0,sort=True) #
tables =  tables.drop(columns=['Flag','Rper']) #included Rper b/c there must be a value

#sort gives it new indices
# tables = tables.drop(['Flag','Rper'], axis=1) #this works too
tables #,type(tables)
```

```python
#REMEMBER TO DROP THIS COLUMN OR ELSE THE TREE HAS THE ANSWERS TO SPLIT ON
#part 1 of 2












#add class to df
tables['Class'] = classes
tables








plt.figure(figsize=(14,8))
plt.scatter(tables['Teff'],tables['Prot'],s=.2,c=tables['Class'],vmin=0,vmax=1)
plt.colorbar().set_label('Class',fontsize=20)
plt.xlabel('Effective Temperature',fontsize=20)
plt.ylabel('Rotation Period',fontsize=20)
plt.xlim(6500,3000);
plt.title('Decision Tree Sample of Stellar Rotation Periods',fontsize=25);






```

```python
#part 2 of 2





tables =  tables.drop(columns=['Class']) #cant give the model the answer
```

```python
# tables.to_csv('mydata.csv',index=False)
# np.save('mydata_labels',classes)
```

```python
######################## 0=valid rp; 1=invalid rp
#and then  create decision tree classifier
```

```python
X = tables #feature data
y = classes #correct answers
```

```python
len(X),len(y)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split



#split df
train_X, test_X, train_y, test_y = train_test_split(X, y, random_state = 11)


# Define your model here
model = DecisionTreeClassifier()

# Fit your model here
fit_model = model.fit(train_X,train_y)
```

```python
train_X
```

```python
from sklearn.metrics import accuracy_score


predicted = fit_model.predict(test_X)
accuracy_score(test_y, predicted)
```

```python
from sklearn import tree


plt.figure(figsize=(20,15))
tree.plot_tree(model.fit(train_X, train_y)); 
print('Tree split on weight which makes sense but next lets try to remove that feature')
```

# Compare after removing weight feature

```python
tables =  tables.drop(columns=['w']) #do a comparison without weight feature

```

```python
X = tables #feature data
y = classes #correct answers
```

```python
#split df
train_X, test_X, train_y, test_y = train_test_split(X, y)#, random_state = 11)


# Define your model here
model = DecisionTreeClassifier()
# Fit your model here
fit_model = model.fit(train_X,train_y)
#Make predicitons
predicted = fit_model.predict(test_X)
#check accuracy
accuracy_score(test_y, predicted)
```

```python
#visualize tree
from sklearn import tree
plt.figure(figsize=(20,15))
tree.plot_tree(model.fit(train_X, train_y)); 
print('this is from the run WITHOUT weight feature')
```
